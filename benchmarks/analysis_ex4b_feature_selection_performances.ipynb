```python
import pickle
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle("dataframes/ex4b_df_selection_performances.pd")
```

```python
comparison = ["feature_importance",
"feature_importance_lda",
"feature_importance_lda_max",
"feature_importance_lr",
"feature_importance_lr_max",
"feature_importance_xgb",
"feature_importance_pca_loadings",
"PCA"]
test_acc = [x + "_test" for x in comparison]
```

```python
def printer(row, name):
    test = name + "_test"
    std = name + "_test_std"
    return "" + ("%.2f" % round(row[test], 2)).lstrip("0") + " $\\pm$" + ("%.2f" % round(row[std], 2)).lstrip("0") + " & "
```

```python
for index, row in df.iterrows():
    str1 = row.dataset.split("_")[0] + " & "    
   # str1 += str(row.n_features - 1) + "& "
    str1 += printer(row,"feature_importance")
    str1 += printer(row,"feature_importance_lda")
    str1 += printer(row,"feature_importance_lda_max")
    str1 += printer(row,"feature_importance_lr")
    str1 += printer(row,"feature_importance_lr_max")
    str1 += printer(row,"feature_importance_xgb")
    str1 += printer(row,"feature_importance_pca_loadings")        
    str1 += printer(row,"PCA")
#    str1 += printer(row,"None")
    str1 = str1[:-2]
    str1 +=  "\\\\"
    print(str1)
           
```

```python
df_div = df[test_acc].div(df[test_acc].max(axis=1),axis=0) - 1
df_div
```

```python
df_div["dataset"] = df["dataset"]
```

```python
def printer2(row, name):
    test = name + "_test"
    return "" + ("%.2f" % round(row[test], 2)).lstrip("0") + " & "
```

```python
for index, row in df_div.iterrows():
    str1 = row.dataset.split("_")[0] + " & "    
   # str1 += str(row.n_features - 1) + "& "
    str1 += printer2(row,"feature_importance")
    str1 += printer2(row,"feature_importance_lda")
    str1 += printer2(row,"feature_importance_lda_max")
    str1 += printer2(row,"feature_importance_lr")
    str1 += printer2(row,"feature_importance_lr_max")
    str1 += printer2(row,"feature_importance_xgb")
    str1 += printer2(row,"feature_importance_pca_loadings")        
    str1 += printer2(row,"PCA")
#    str1 += printer(row,"None")
    str1 = str1[:-2]
    str1 +=  "\\\\"
    str1 = str1.replace("-0","-").replace("-.00",".00").replace(".00", "\\cellcolor{custom-grey}.00")
    print(str1)
```

```python
average_percentage_diff = (df[test_acc].div(df[test_acc].max(axis=1),axis=0) - 1).mean()
average_percentage_diff
```

```python
str1 = "average & "
for element in round(average_percentage_diff,2):
    str1 += str(element) + " & "
    str1 = str1.replace("-0","-")
print(str1)
```
